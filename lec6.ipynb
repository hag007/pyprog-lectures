### Some available operator overloads

By defining some special methods, you can specify the behavior of operators on user defined types

### Operator overload: `+` implementation

$$
a/b+c/d=(a\cdot b)/(b\cdot d)
$$

````python
class Rational:
    def __init__(self, n, d=1):
        """n: numerator
        d: denominator"""
        if d == 0:
            print("denominator is set to 1, cannot be zero")
            d = 1
        self.numer = n
        self.denom = d
    
    def __repr__(self):
        return str(self.numer) + '/' + str(self.denom)
    
    ### New code here ###
    def __add__(self, other):
        return Rational(self.numer * other.denom + \
        other.numer * self.denom, \
        self.denom * other.denom)
    #####################
    
q = Rational(1,2)
r = Rational(2,3)
q+r
````

### Customized `max`

````python
class Beatle:
    def __init__(self, name, popularity, awesomeness):
        self.name = name
        self.popularity = popularity
        self. awesomeness = awesomeness
    def __repr__(self):
        return self.name + str((self.popularity, self. awesomeness))

beatles = []
beatles.append(Beatle('John', 10, 7))
beatles.append(Beatle('Paul', 9, 8))
beatles.append(Beatle('George', 7, 10))
beatles.append(Beatle('Ringo', 5, 5))

print(beatles)
````

How can we find out what's is the most popular Beatle and what's the most awesome one?

### Finding an object with maximal attribute value

- We can add `__lt__`  implementation and use `max`
- Caveat:
    - When we implement __lt__ we'll have to choose a single comparison criterion: either popularity or awesomeness



### Using max with the key argument 
`max(iterable[, key])`
- key: an optional argument for the max function. If specified, key is applied on each element of iterable.
- max returns the element elem s.t. `key(elem)` is the maximum value among the other elements in iterable.

### Finding max element in our Beatles example 

We can implement helper-functions that will help us to compare Beatles according to different criteria

````python
def popularity_key(beatle):
    return beatle.popularity 
def awesomeness_key(beatle):
    return beatle.awesomeness
````

````python
print(beatles)
````

````python
print(max(beatles, key = popularity_key))
print(max(beatles, key = awesomeness_key))
````

### `key` argument also exists in other order-based functions (`min`, `sort`, etc.)

````python
print(min(beatles, key = awesomeness_key))
````

````python
print(sorted(beatles, key = awesomeness_key))
````

````python
class Ingredient:
    def __init__(self, name, calories, weight=100):
        self.name = name
        self.calories = calories
        if type(weight) in [int, float] and weight > 0:
            self.weight = weight
        else:
            self.weight = 100
            print('weight error')
             
    def calc_calories(self):
        return int(self.weight / 100 * self.calories)
    
    def __repr__(self):
        return f"{self.name}: {self.calc_calories()} calories"
````

````python
lettuce = Ingredient('Lettuce', 17, 32)
egg = Ingredient ('Egg', 60, 134)
mushroom = Ingredient ('Mushroom', 25, 22)
onion = Ingredient ('Onion', 25, 200)
print(lettuce.calc_calories())
print(egg)
print(mushroom)
print(onion)
lettuce = Ingredient('Lettuce', 17, -4)
lettuce = Ingredient('Lettuce', 17, 'X')
````

````python
class Recipe:

    def __init__(self, name, time, ing_lst):
        self.name = name
        self.time = time
        if ing_lst:
            self.ing_lst = ing_lst
        else:
            print('ingridients error')

    def calc_calories(self):
        return sum([amount * ing.calc_calories() for amount, ing in self.ing_lst])

    def __repr__(self):
        result = f"{self.name}, {self.time} minutes \n"
        for amount, ing in self.ing_lst:
            result += f"{ing} * {amount}\n"
        result += f"total = {self.calc_calories()}"
        return result
````
